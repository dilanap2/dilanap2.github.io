```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
from vega_datasets import data
df = pd.read_csv('nba_stats.csv')
df = df[df["MP"] >= 20]
df
```

```python
df.columns
```

The name of the dataset is 2022-23 NBA Player Stats. I got this data from the https://www.basketball-reference.com/ website. https://www.basketball-reference.com/leagues/NBA_2023_per_game.html. on the webiste, it says "When using SR data, please cite us and provide a link and/or a mention." The website has a page with rules for using their data here: https://www.sports-reference.com/data_use.html. After reading through it, I know I will not break the rules as it mostly refers to people trying to profit off of their data. The CSV is 96KB. 

```python
position_rpg = df.groupby('Pos')['TRB'].mean().sort_index()
plt.figure(figsize=(8, 6))
position_rpg.plot(kind='bar', color='skyblue')
plt.title('Average Rebounds per Game (RPG) by Position (Min 20 MPG')
plt.xlabel('Position')
plt.ylabel('Average RPG')
plt.xticks(rotation=0) 
plt.show()
```

```python
position_apg = df.groupby('Pos')['AST'].mean().sort_index()
plt.figure(figsize=(8, 6))
position_apg.plot(kind='bar', color='skyblue')
plt.title('Average Assists per Game (APG) by Position (Min 20 MPG)')
plt.xlabel('Position')
plt.ylabel('Average PPG')
plt.xticks(rotation=0) 
plt.show()
```

```python
position_ppg = df.groupby('Pos')['PTS'].mean().sort_index()
plt.figure(figsize=(8, 6))
position_ppg.plot(kind='bar', color='skyblue')
plt.title('Average Points per Game (PPG) by Position (Min 20 MPG)')
plt.xlabel('Position')
plt.ylabel('Average PPG')
plt.xticks(rotation=0) 
plt.show()
```

```python
alt.Chart(df).mark_circle(size=60).encode(
    x='Age',
    y='PTS',
    color='Tm',
    tooltip=['Player','PTS', 'TRB']
).interactive()
```
