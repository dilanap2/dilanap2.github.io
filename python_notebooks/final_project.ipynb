```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
from vega_datasets import data
df = pd.read_csv('nba_stats.csv')
df = df[df["MP"] >= 20]
df
```

```python
position_rpg = df.groupby('Pos')['TRB'].mean().sort_index()
plt.figure(figsize=(8, 6))
position_rpg.plot(kind='bar', color='skyblue')
plt.title('Average Rebounds per Game (RPG) by Position (Min 20 MPG')
plt.xlabel('Position')
plt.ylabel('Average RPG')
plt.xticks(rotation=0) 
plt.show()
```

```python
position_apg = df.groupby('Pos')['AST'].mean().sort_index()
plt.figure(figsize=(8, 6))
position_apg.plot(kind='bar', color='skyblue')
plt.title('Average Assists per Game (APG) by Position (Min 20 MPG)')
plt.xlabel('Position')
plt.ylabel('Average PPG')
plt.xticks(rotation=0) 
plt.show()
```

```python
position_ppg = df.groupby('Pos')['PTS'].mean().sort_index()
plt.figure(figsize=(8, 6))
position_ppg.plot(kind='bar', color='skyblue')
plt.title('Average Points per Game (PPG) by Position (Min 20 MPG)')
plt.xlabel('Position')
plt.ylabel('Average PPG')
plt.xticks(rotation=0) 
plt.show()
```

These three graphs that show the average points per game per position, the average assists per game per position, and the average rebounds per game per position for the entire NBA can be useful for coaches, scouts, and analysts to gain insights into the league's overall performance trends. By analyzing the averages of these metrics across different positions, it becomes possible to identify general areas of strength or weakness within the league. For example, if the averages of points per game are highest for shooting guards and point guards, coaches and scouts may focus on recruiting or developing players who excel in those positions. Similarly, if the averages of assists per game are highest for point guards and small forwards, coaches may adjust their strategies to rely more heavily on players in those positions to create scoring opportunities for the team. These graphs can also be used to identify trends over time, such as changes in the average rebounds per game by position from one season to the next. Overall, these graphs provide a valuable tool for evaluating performance trends across the entire NBA and can help inform strategic decisions for teams, scouts, and analysts.
